# Introduction/Business Problem

Arlington, Texas, with approximately four hundred thousand residents, is the largest city in the United States by population without any mass transportation. It's estimated that 82% of Arlington's residents over 16 drive to work alone. This high traffic volume burdens the community with air pollution, noise pollution, some of the highest car insurance premiums in the country, more frequent needs for road construction, and potentially smaller profits for Arlington's largest industry: tourism.  The driving burden on the community could be diminshed with efficient public transport. I seek to use Fouresquare location data to identify some potential placements for fixed route bus stops.

# Data

I will be using Foursquare to cluster neighborhoods of the city based on peak times in activity throughout the week to identify potential trends in transporotation needs. I will also be looking at city accident data to determine areas of the city with higher traffic burdens.
